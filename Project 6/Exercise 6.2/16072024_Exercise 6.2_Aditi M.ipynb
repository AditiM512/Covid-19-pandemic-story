```python
df_pivot['Vaccinated Percentage'] = (df_pivot['Max of people_vaccinated'] / df_pivot['Max of population']) * 100
```

```python
df_pivot.head()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df_pivot['Vaccinated Percentage'], bins=10, edgecolor='black')
plt.xlabel('Vaccinated Percentage (%)')
plt.ylabel('Frequency')
plt.title('Histogram of Vaccinated People Percentage')
plt.grid(True)
plt.show()
```

```python
df_pivot.loc[df_pivot['Vaccinated Percentage'] < 30, 'Vaccinated_Range'] = 'Low percentage'
df_pivot.loc[(df_pivot['Vaccinated Percentage'] >= 30) & (df_pivot['Vaccinated Percentage'] < 50), 'Vaccinated_Range'] = 'Mid-range'
df_pivot.loc[df_pivot['Vaccinated Percentage'] >= 50, 'Vaccinated_Range'] = 'High percentage'
```

```python
df_pivot['Vaccinated_Range'].value_counts(dropna = False)
```

```python
df_pivot['total_case Percentage'] = (df_pivot['Max of total_cases'] / df_pivot['Max of population']) * 100
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
Categorical_plot = sns.catplot(x="total_case Percentage", y="Vaccinated Percentage", hue="Vaccinated_Range", data=df_pivot)
```
