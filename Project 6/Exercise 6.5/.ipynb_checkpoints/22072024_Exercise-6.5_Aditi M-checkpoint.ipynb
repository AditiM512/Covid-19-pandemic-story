```python
file_path = r'C:\Users\aditi\Desktop\Data Analytics_Career Foundry\Data Analytics_Immersion\Project 6\Pivoted_dataset.xlsx'
```

```python
df = pd.read_excel(file_path)
```

```python
df=df[['Max of total_cases', 'Max of total_deaths', 'Max of reproduction_rate', 'Max of people_vaccinated', 'Max of people_fully_vaccinated', 'Average of aged_65_older (in %)', 'Average of gdp_per_capita (in USD)', 'Average of cardiovasc_death_rate (per 100,000)', 'Average of diabetes_prevalence (%)', 'Average of female_smokers (%)', 'Average of male_smokers (%)', 'Average of life_expectancy (in years)', 'Max of population']]
```

```python
df['Total_cases_Percentage'] = (df['Max of total_cases'] / df['Max of population']) * 100
```

```python
df['Total_deaths_Percentage'] = (df['Max of total_deaths'] / df['Max of population']) * 100
```

```python
df.head()
```

The Elbow Technique (From Chat GpT since method in the exercise was generating warning "FutureWarning: The default value of `n_init` will change from 10 to 'auto' in 1.4. Set the value of `n_init` explicitly to suppress the warning")

```python
range_n_clusters = range(1, 11) # Defines the range of potential clusters in the data.
# List to store the WCSS values
wcss = []
```

```python
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method graph
plt.plot(range_n_clusters, wcss, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()
```

# k-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 2)
```

```python
# Fit the k-means object to the data.

kmeans.fit(df)
```

```python
df['clusters'] = kmeans.fit_predict(df)
```

```python
df.head()
```

```python
df['clusters'].value_counts()
```

```python
# Plot the clusters for the "Average of aged_65_older (in %)" and "Total_deaths_Percentage" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Average of aged_65_older (in %)'], y=df['Total_deaths_Percentage'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Average of aged_65_older (in %)') # Label x-axis.
plt.ylabel('Total_deaths_Percentage') # Label y-axis.
plt.show()
```

```python
# Filter and view the data points in cluster '1'
cluster_1_data = df[df['clusters'] == 1]
```

```python
print("\nData points in cluster '1':")
print(cluster_1_data)
```

```python
df.loc[df['clusters'] == 1, 'Cluster'] = 'Orange'
df.loc[df['clusters'] == 0, 'Cluster'] = 'Blue'
```

```python
df_agg=df.groupby('Cluster').agg({'Max of population':['mean', 'median'], 
                         'Total_cases_Percentage':['mean', 'median'], 
                         'Total_deaths_Percentage':['mean', 'median']})
```

# Filtering out India and China to see new clusters.

```python
populations_to_remove = [1425887360, 1417173120]

df_filtered = df[~df['Max of population'].isin(populations_to_remove)]
```

```python
df_filtered.head()
```

```python
df_reset = df_filtered.reset_index(drop=True)
```

```python
df_reset.head()
```

```python
df_reset = df_reset.drop(columns=['clusters', 'Cluster'])
```

```python
range_n_clusters = range(1, 11) # Defines the range of potential clusters in the data.
# List to store the WCSS values
wcss = []
```

```python
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
    kmeans.fit(df_reset)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method graph
plt.plot(range_n_clusters, wcss, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 3)
```

```python
# Fit the k-means object to the data.

kmeans.fit(df_reset)
```

```python
df_reset['Clusters'] = kmeans.fit_predict(df_reset)
```

```python
df_reset['Clusters'].value_counts()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_reset['Average of aged_65_older (in %)'], df_reset['Total_deaths_Percentage'], c=df_reset['Clusters'], cmap='viridis', s=100)

# Adding a color bar
plt.colorbar(scatter, label='Cluster')

# Adding labels and title
plt.xlabel('Average of aged_65_older (in %)')
plt.ylabel('Total_deaths_Percentage')
plt.title('Scatter Plot of Clusters')

# Show plot
plt.show()
```

```python
df_reset.loc[df_reset['Clusters'] == 1, 'Cluster'] = 'Green'
df_reset.loc[df_reset['Clusters'] == 0, 'Cluster'] = 'Purple'
df_reset.loc[df_reset['Clusters'] == 2, 'Cluster'] = 'Yellow'
```
