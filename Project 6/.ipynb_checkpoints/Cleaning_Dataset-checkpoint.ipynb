```python
# Import dataset
df_dataset = pd.read_csv(os.path.join(path, 'owid-covid-data.csv'))
```

```python
selected_columns = ['iso_code', 'continent', 'location', 'date', 'total_cases', 'total_deaths', 'reproduction_rate', 'icu_patients', 'hosp_patients', 'total_tests', 'people_vaccinated', 'people_fully_vaccinated', 'aged_65_older', 'gdp_per_capita', 'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers', 'male_smokers', 'life_expectancy', 'population']
df_new_dataset = df_dataset[selected_columns]
```

```python
all_null_rows = df_new_dataset.isna().all(axis=1)
```
